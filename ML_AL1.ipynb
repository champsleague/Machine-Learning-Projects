```python
%matplotlib inline
import numpy as np
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn.datasets import make_blobs

# determines the center points
centers = [[3,3,3],[4,5,5],[3,10,10]]
X, _ = make_blobs(n_samples = 700, centers = centers, cluster_std = 0.5)

# print as the scatter point
plt.scatter(X[:,0],X[:,1])
plt.show()


ms = MeanShift()
# print as applying the mean shift
ms.fit(X)
labels = ms.labels_

# print the output of cluster centers
cluster_centers = ms.cluster_centers_
print('Output  \n',cluster_centers)

# calculate the estimated clusters
n_clusters_ = len(np.unique(labels))
print("Estimated clusters:", n_clusters_)
colors = 10*['r.','g.','b.','c.','k.','y.','m.']

# set the range of X
for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 3)
plt.scatter(cluster_centers[:,0],cluster_centers[:,1],
    marker=".",color='k', s=20, linewidths = 5, zorder=10)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# define the position for the cluster in 3D graph
position = [[3,3,5],[2,2,2],[1,5,5]]
X,_=make_blobs(n_samples=150,centers=position,cluster_std=0.70)

# apply the meanshift and get the coordinates of the clusters
ms = MeanShift()
ms.fit(X)
cluster_centers = ms.cluster_centers_

# Plotting the 3D plot of the clusters with its centered marked and represent the output as 3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], marker='o')

# mark 'x' as the cluster centers
ax.scatter(cluster_centers[:,0], cluster_centers[:,1], 
 cluster_centers[:,2], marker='x', color='blue', s=300, linewidth=5,zorder=10)
plt.show()
```
